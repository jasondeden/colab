# Bloom Hello World in Easily Reproducible Environment

## This notebook was designed to perform early POC testing on Bloom in an environment that could be easily replicated.

```python
import transformers
```

```python
from transformers import BloomForCausalLM
```

```python
from transformers import BloomTokenizerFast
```

```python
model = BloomForCausalLM.from_pretrained("bigscience/bloom-350m") #Had to use 350m instead of 1b3 due to resource limitations - ok for testing
```

```python
tokenizer = BloomTokenizerFast.from_pretrained("bigscience/bloom-350m")
```

```python
prompt = "I really wish I could" #This is the text string that the LLM will finish
result_length = len(prompt.split()) +15 #This determines the number of words to follow the prompt
inputs = tokenizer(prompt, return_tensors="pt") 
```

```python
# Greedy Search
print(tokenizer.decode(model.generate(inputs["input_ids"], 
                       max_length=result_length
                      )[0]))
```

```python
# Beam Search
print(tokenizer.decode(model.generate(inputs["input_ids"],
                       max_length=result_length, 
                       num_beams=2, 
                       no_repeat_ngram_size=2,
                       early_stopping=True
                      )[0]))
```

```python
# Sampling Top-k + Top-p
print(tokenizer.decode(model.generate(inputs["input_ids"],
                       max_length=result_length, 
                       do_sample=True, 
                       top_k=50, 
                       top_p=0.9
                      )[0]))
```
